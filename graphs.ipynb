# Plotting Multiple Charts in a Grid

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Dataset for 1st Graph
years =range(2000,2006)
apples = [0.35,0.6,0.9,0.8,0.65,0.8]
oranges =[0.4,0.8,0.9,0.7,0.6,0.8]
```

```python
# Dataset from Seaborn for 2nd & 3rd Graph
flower = sns.load_dataset('iris')

setosa = flower[flower.species == 'setosa']
vc = flower[flower.species == 'versicolor']
virginica = flower[flower.species == 'virginica']
```

```python
# Dataset from Seaborn for 4th Graph
tips = sns.load_dataset('tips')
```

```python
# Dataset from Seaborn for 5th Graph
flights = sns.load_dataset('flights').pivot(index='month',columns='year',values='passengers')
```

```python
# Image for 6th Graph
from PIL import Image
img = Image.open('funny.jpg')
```

```python
fig,axes = plt.subplots(2,3,figsize=(16,8))
plt.tight_layout(pad=3)


# 1st
axes[0,0].plot(years,apples,'s-r')
axes[0,0].plot(years,oranges,'o--b')
axes[0,0].set_xlabel('Year')
axes[0,0].set_ylabel('Yields (tons per hectare)')
axes[0,0].legend(['Apples','Oranges'])
axes[0,0].set_title('Crop Yields in Kanto')

#2nd
axes[0,1].set_title('Sepal Length vs. Sepal Width')
sns.scatterplot(y=flower.sepal_width,
                x=flower.sepal_length,
                hue= flower.species,
                ax=axes[0,1])

# 3rd
axes[0,2].set_title('Distribution of Sepal Width')
axes[0,2].hist([setosa.sepal_width,vc.sepal_width,virginica.sepal_width],
               bins= np.arange(2,5,0.25),
               stacked = True,
               rwidth= 0.98)
axes[0,2].legend(['Setosa','Versicolor','Virginica'])

# 4th
axes[1,0].set_title('Restaurant Bills')
sns.barplot(x='day',y='total_bill',hue='sex',data=tips,ax=axes[1,0])

# 5th
axes[1,1].set_title('Flight Traffic')
sns.heatmap(flights,cmap='Blues',ax=axes[1,1])

# 6th
axes[1,2].set_title('Matplotlib Meme')
axes[1,2].imshow(img)
axes[1,2].grid(False)
axes[1,2].axis('on')

plt.show()
```
